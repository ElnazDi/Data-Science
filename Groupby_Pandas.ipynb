```
ft = data_1['Country'] == 'India'
```

```
data_1.loc[ft]['LanguageWorkedWith'].str.contains('Python').sum()
```

```
g = data_1.groupby('Country')
```

```
people_python =g['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
people_python
```

```
people = data_1['Country'].value_counts()
people
```

```
python_df = pd.concat([people, people_python], axis='columns', sort=False)
python_df
```

```
python_df.rename(columns={'Country': 'number_of_respondent', 'LanguageWorkedWith': 'Number_of_people_know_Python'}, inplace=True)
```

```
python_df['pct'] = (python_df['Number_of_people_know_Python'] / python_df['number_of_respondent']) *100
python_df
```

```
python_df.sort_values(by='pct', ascending=False, inplace=True)
```
