```
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from google.colab import files
```

```
data = pd.read_csv('ign_scores.csv', index_col='Platform')
data.head()
```

```
data   #Since our data set is small we can print all in once
```

```
list(data.columns)
```

```
plt.figure(figsize=(30,6))
sns.barplot(x=data.index , y=data['Racing'])     # since in impoting data we wrote index_col='Platform'  we can not use x=data['Platform']
plt.ylabel('')                                   #We do not want to write anything
plt.title('Average Score for Racing Games, by Platform')
```

```

# Better to visualize data in this way, it is compacted

plt.figure(figsize=(8,6))
sns.barplot(x=data['Racing'], y=data.index )     # since in impoting data we wrote index_col='Platform'  we can not use y=data['Platform']
plt.xlabel('')                                   #We do not want to write anything
plt.title('Average Score for Racing Games, by Platform')
```

```
plt.figure(figsize=(10,10))
sns.heatmap(data=data, annot= True)    # If we do not write annot=True , numbers will not be shown on each squars
plt.xlabel("Genre")
plt.title('Average Game Score, by Platform and Genre')
```
