```
import pandas as pd

# we wrote full because there are all features in them

X_full_seen_train = pd.read_csv('train.csv', index_col='Id')
X_full_unseen_test = pd.read_csv('test.csv', index_col='Id')
```

```
y = X_full_seen_train.SalePrice
print(y.head())
```

```
# We want only some features in X_full_seen_train not all of them

features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full_seen_train[features].copy()
X_test = X_full_unseen_test[features].copy()  # Not important for spliting
```

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y, train_size= 0.8, test_size= 0.2, random_state=0)
```

```
from sklearn.ensemble import RandomForestRegressor

model_1 = RandomForestRegressor(n_estimators=50 ,random_state=0)
model_2 = RandomForestRegressor(n_estimators=100 ,random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='mae', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20 ,random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7 ,random_state=0)

models = [ model_1, model_2, model_3, model_4, model_5 ]
```

```
from sklearn.metrics import mean_absolute_error

def score_model(model, X_t = X_train, X_v = X_val, y_t = y_train, y_v = y_val):
  model.fit(X_t, y_t)
  prediction = model.predict(X_v)
  mae = mean_absolute_error(y_v, prediction)
  return mae
```

```
for i in range(0, len(models)):         #Remmeber range(3,5) numbers between 3 and 5 but not including 5  or range(5) => 0,1,2,3,4
  model_mae = score_model(models[i])    #len(models)=5 so it is range(0,5) => 0,1,2,3,4
  print("Model %d \t\t MAE: %d" %(i+1, model_mae))
```

```
best_model = model_3
```

```
my_chosen_model = best_model
my_chosen_model.fit(X,y) #When we have best model we can train it with all data
prediction_unseen= my_chosen_model.predict(X_test) #here we do not use X_val !!! because we want to use it for a real unseen data
print(prediction_unseen)
```

```
output = pd.DataFrame({'Id': X_test.index , 'SalePrice': prediction_unseen})
output.to_csv('predictions.csv', index=False)
```
