```
import pandas as pd
from matplotlib import pyplot as plt
```

```
data=pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv')
data.head()
```

```
data.columns
```

```
ages=data['Age']
dev_salaries= data['All_Devs']
py_salaries= data['Python']
js_salaries= data['JavaScript']

plt.plot(ages, dev_salaries, label='Dev Salaries', color='#444444', linestyle='--')
plt.plot(ages, py_salaries, label= 'Py Salaries')


plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by their Ages')

plt.legend()
plt.tight_layout()
plt.show()
```

```
ages=data['Age']
dev_salaries= data['All_Devs']
py_salaries= data['Python']
js_salaries= data['JavaScript']

plt.plot(ages, dev_salaries, label='Dev Salaries', color='#444444', linestyle='--')
plt.plot(ages, py_salaries, label= 'Py Salaries')


plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by their Ages')

plt.fill_between(ages, py_salaries)

plt.legend()
plt.tight_layout()
plt.show()
```

```
ages=data['Age']
dev_salaries= data['All_Devs']
py_salaries= data['Python']
js_salaries= data['JavaScript']

plt.plot(ages, dev_salaries, label='Dev Salaries', color='#444444', linestyle='--')
plt.plot(ages, py_salaries, label= 'Py Salaries')

#overall_median = 57287


plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by their Ages')

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate= True, alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate= True, color='red', alpha=0.25, label='Below Avg')


plt.legend()
plt.tight_layout()
plt.show()
```
