# categorical variable : with three approaches 



1.    **Drop Categorical Variables:**

simply remove them from the dataset. This approach will only work well if the columns did not contain useful information.

2.  **Label Encoding :** Ordinal variables 

"Never" (0) < "Rarely" (1) < "Most days" (2) < "Every day" (3)

For tree-based models (like decision trees and random forests), you can expect label encoding to work well with ordinal variables.

3.   **One  Hot Encoding:**  Nominal variables 

does not perform well if the categorical variable takes on a large number of values

**We can see lables in X_train and X_valid are not same, This will cause error**

Fitting a label encoder to a column in the training data creates a corresponding integer-valued label for each unique value that appears in the training data. In the case that the validation data contains values that don't also appear in the training data, the encoder will throw an error, because these values won't have an integer assigned to them. Notice that the 'Condition2' column in the validation data contains the values 'RRAn' and 'RRNn', but these don't appear in the training data -- thus, if we try to use a label encoder with scikit-learn, the code will throw an error.






This is a common problem that you'll encounter with real-world data, and there are many approaches to fixing this issue. For instance, you can write a custom label encoder to deal with new categories. The simplest approach, however, is to drop the problematic categorical columns.



```
# We want to seperate Good lables (sames) from Bad lables (differents) in both test and train

# Select all columns with categorical data
#Select all columns with Good lables (same in both train and val)
# Bad_lables = All - Good

all_categorical_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']
good_lable_cols = [col for col in all_categorical_cols  if set(X_train[col]) ==  set(X_valid[col])]
bad_lable_cols = list(set(all_categorical_cols) - set(good_lable_cols))

print("Good Lables are: \n", good_lable_cols)
print("\nBad Lables are: \n", bad_lable_cols)

```

```
# Now we need to drop these bad lables from dataset

from sklearn.preprocessing import LabelEncoder

# Drop categorical columns that will not be encoded
X_train_GL = X_train.drop(bad_lable_cols, axis=1)
X_valid_GL = X_valid.drop(bad_lable_cols, axis=1)

lable_encoder = LabelEncoder()
for col in set(good_lable_cols):
  X_train_GL[col] = lable_encoder.fit_transform(X_train[col])
  X_valid_GL[col] = lable_encoder.transform(X_valid[col])
```

```
# Get score for this approach

print("Mean Absolute Error in Second Approach: \n")

mae = get_score(X_train_GL, X_valid_GL, y_train, y_valid)
print(mae)
```

# Step 4: One-hot encoding

In this step, you'll experiment with one-hot encoding.  But, instead of encoding all of the categorical variables in the dataset, you'll only create a one-hot encoding for columns with cardinality less than 10.

Run the code cell below without changes to set `low_cardinality_cols` to a Python list containing the columns that will be one-hot encoded.  Likewise, `high_cardinality_cols` contains a list of categorical columns that will be dropped from the dataset.

```
# Columns that will be one-hot encoded
low_cardinality_cols = [col for col in all_categorical_cols if X_train[col].nunique() < 10]

# Columns that will be dropped from the dataset
high_cardinality_cols = list(set(all_categorical_cols)-set(low_cardinality_cols))

print('Categorical columns that will be one-hot encoded:', low_cardinality_cols)
print('\nCategorical columns that will be dropped from the dataset:', high_cardinality_cols)
```

We use the OneHotEncoder class from scikit-learn to get one-hot encodings. There are a number of parameters that can be used to customize its behavior.

 **handle_unknown='ignore'**

to avoid errors when the validation data contains classes that aren't represented in the training data, and


***setting sparse=False ***

ensures that the encoded columns are returned as a numpy array (instead of a sparse matrix).

```
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[low_cardinality_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[low_cardinality_cols]))

# One-hot encoding removed index; put it back
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(all_categorical_cols, axis=1)
num_X_valid = X_valid.drop(all_categorical_cols, axis=1)

# Add one-hot encoded columns to numerical features
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

```
print("MAE from Approach 3 (One-Hot Encoding):") 
print(get_score(OH_X_train, OH_X_valid, y_train, y_valid))
```
