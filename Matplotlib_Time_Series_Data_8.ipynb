```
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y, linestyle='solid')



plt.gcf().autofmt_xdate()  #Italic Dates ---- get current format=gcf  fmt=format

date_format =  mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)   # get current axis


#plt.ylabel('Closing Price')
#plt.xlabel('Date')
#plt.title('Bitcoin Prices')

plt.tight_layout()
plt.show()
```

```
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv' )
data.head()
```

```
price_date= data['Date']
price_close= data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')



plt.gcf().autofmt_xdate()  #Italic Dates ---- get current format=gcf  fmt=format



#----------------------------------#
plt.ylabel('Closing Price')
plt.xlabel('Date')
plt.title('Bitcoin Prices')

plt.tight_layout()
plt.show()
```
